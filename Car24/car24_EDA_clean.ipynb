```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("car_data_car24.csv")
data
```

```python
data.duplicated().sum()
```

```python
data=data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
data.info()
```

```python
data
```

```python
# Remove 'k km', convert to float
data["km_travelled"] = (
    data["km_travelled"]
    .str.replace("km","", regex=False)
)
```

```python
data
```

```python
data["discounted_final_price"] = data["discounted_final_price"].str.replace(" lakh", "L", regex=False)
```

```python
data
```

```python
# Drop rows where discounted_final_price is exactly "Account"
data = data[data["discounted_final_price"].str.strip() != "Account"]
```

```python
data
```

```python
data["discounted_final_price"] = (
    data["discounted_final_price"]
    .str.replace("₹", "", regex=False)
    .str.replace("L", "", regex=False)
)
```

```python
data
```

```python
data["discounted_final_price"].info()
```

```python
data["discounted_final_price"] = pd.to_numeric(data["discounted_final_price"], errors="coerce")
```

```python
data["discounted_final_price"].info()
```

```python
data["discounted_final_price"].isnull().sum()
```

```python
data["discounted_final_price"] = data["discounted_final_price"].fillna(data["discounted_final_price"].median())
```

```python
data["discounted_final_price"].isnull().sum()
```

```python
data
```

```python
data["original_price"] = data["original_price"].str.replace("₹", "", regex=False)
```

```python
data["discounted_final_price"] = data["discounted_final_price"].astype(str) + "L"
```

```python
data
```

```python
# Extract 4-digit year from car_name and create manufacturing_year column
data["manufacturing_year"] = data["car_name"].str.extract(r"(\d{4})").astype(int)
```

```python
data
```

```python
# Remove the 4-digit year from car_name
data["car_name"] = data["car_name"].str.replace(r"^\d{4}\s*", "", regex=True)
```

```python
data
```

```python
import numpy as np
data["selling_Location"] = data["selling_Location"].replace("Cars listed by CARS24", np.nan)
```

```python
data
```

```python
data["selling_Location"] = data["selling_Location"].fillna(data["selling_Location"].mode()[0])
```

```python
data
```

```python
# Extract city name from selling_Location
data["city"] = data["selling_Location"].str.extract(r",\s*([^,]+)$")
```

```python
# Remove the last comma and city name
data["selling_Location"] = data["selling_Location"].str.replace(r",\s*[^,]+$", "", regex=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data["discount"] = data["original_price"].str.replace("L", "", regex=False).astype(float) - \
                   data["discounted_final_price"].str.replace("L", "", regex=False).astype(float)
```

```python
# Remove 'L' and convert to float
data["original_price"] = data["original_price"].str.replace("L", "", regex=False).astype(float)
data["discounted_final_price"] = data["discounted_final_price"].str.replace("L", "", regex=False).astype(float)
```

```python
data
```

```python
data["discount_percent"] = (
    (data["original_price"] - data["discounted_final_price"]) / data["original_price"]
) * 100
```

```python
data
```

```python
data["discount_percent"] = data["discount_percent"].astype(int)
```

```python
data
```

```python
data["discount_percent"] = data["discount_percent"].astype(str) + "%"
```

```python
data["original_price"] = data["original_price"].astype(str) + "L"
data["discounted_final_price"] = data["discounted_final_price"].astype(str) + "L"
```

```python
data
```

```python
data.drop(columns=["discount"], inplace=True)
```

```python
data
```

```python
data["city"] = data["city"].fillna(data["city"].mode()[0])
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data
```

```python
data.to_csv("car24_final_dataset.csv")
```
